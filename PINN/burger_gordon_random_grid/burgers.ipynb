```python
import burgers_spectral_solution_burkardt as burgers
import matplotlib.pyplot as plt
import numpy as np
from DictX import *
```

```python
dct_ex = DictX()
dct_ex.nu = 2.e-2
dct_ex.nx = 81
dct_ex.nt = 51
dct_ex.xlo = -1.
dct_ex.xhi = 1.
dct_ex.tlo = 0.
dct_ex.thi = 4.
dct = burgers.burgers_viscous_time_exact1_test01_no_output(**dct_ex)

x, t = np.meshgrid(dct.x, dct.t, indexing='ij')
print(x.shape, t.shape)

fig, axes = plt.subplots(1, 2, figsize=(25, 8))
ax = axes[0]
ax.contourf(x, t, dct.u)
ax = axes[1]
for j in range(0, dct_ex.nx, 5):
    ax.plot(x[:,0], dct.u[:, j], label='t = %g' % dct.t[j])
ax.grid(True, alpha=0.5)
ax.legend()
```
